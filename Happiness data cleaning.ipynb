```python
import pandas as pd


ess_data = pd.read_csv("ESS9e03_1.csv")
```

### Checking for null values in the day columns

```python
ess_data[['inwdds','inwdde']].isna().sum()
```

### Converting the day columns to integers rather than floats and filling the null_values with -1

```python
ess_data['day_of_month_start'] = ess_data['inwdds'].fillna(-1).astype(int)
ess_data['day_of_month_end'] = ess_data['inwdde'].fillna(-1).astype(int)
```

### Selecting the columns I would like to use for my analysis

```python
cols = ['idno','cntry','eisced','rlgdgr','agea','blgetmg','happy','lrscale','marsts','hinctnta','inwyys','inwmms','day_of_month_start','day_of_month_end','inwmme','inwyye']


a = ess_data[cols]
```

```python
a
```

### Removing the rows with -1(null) values from the dataset

```python
filt = ~a.iloc[:,2:-2].lt(0)
non_null = filt[filt.all(axis =1)].index.tolist()
```

```python
date_cl_a = a.iloc[non_null,:]
```

### Connecting country abbreviation to country name

```python
abbrev = list(date_cl_a['cntry'].unique())

abbrev
```

```python
countries = ['Austria','Belgium','Bulgaria','Switzerland','Cyprus','Czech Republic','Germany','Denmark','Estonia',
            'Spain','Finland','France','United Kingdom','Croatia','Hungary','Ireland','Iceland','Italy','Lithuania',
             'Latvia','Montenegro','Netherlands','Norway','Poland','Portugal','Serbia','Sweden','Slovenia','Slovakia']
```

```python
len(abbrev) == len(countries)
```

```python
country_full = dict(zip(abbrev,countries))
```

```python
date_cl_a['full_country'] = date_cl_a['cntry'].replace(country_full)
```

```python
date_cl_a
```

## Putting full_country next to cntry

```python
a3 = date_cl_a.reindex(columns = ['idno', 'cntry','full_country', 'eisced', 'rlgdgr', 'blgetmg', 'happy',
        'hinctnta','lrscale','marsts','agea','inwyys', 'inwmms',
       'day_of_month_start', 'day_of_month_end', 'inwmme', 'inwyye',
       ])
```

```python
a3
```

```python
a3.isna().sum()

```

### Getting rid of values greater than 10 since they mean a missing value in the data (check survey documentation for this)

```python
mask = ~a3.iloc[:,3:-7].gt(10)

i = mask[mask.all(axis =1)].index.tolist()
```

```python
(~mask).sum() # most of the missing values come from the marital status question
```

### Decided to drop marital status as a field

```python

a4 = a3.drop(columns = "marsts")
```

```python
mask2 = ~a4.iloc[:,3:-7].gt(10)

i = mask2[mask2.all(axis =1)].index.tolist()
```

```python
(~mask2).sum()
```

### I just filtered out all values greater than 10, essentially where there was no response

```python

no_10 = a4.loc[i]
```

```python
len(no_10)
```

###  This particular question has missing values above the number 6

```python

no_10.loc[:,['blgetmg']].gt(6).sum()
```

```python
x = ~no_10.loc[:,['blgetmg']].gt(6)

i = x[x.all(axis =1)].index.tolist()
a5 = no_10.loc[i]
```

### The age question has missing values when its value is 999 (again check documentation)

```python
clean_data = a5[~(a5.agea == 999)]
```

```python
clean_data[clean_data.agea != 999]
```

```python
clean_data.isna().sum()
```
